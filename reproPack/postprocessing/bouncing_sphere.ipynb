```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# load data from simulation
par1 = np.loadtxt('data/two_phase_falling_sphere/position_falling_sphere_water_air.txt')
par1vel = np.loadtxt('data/two_phase_falling_sphere/velocity_falling_sphere_water_air.txt')
data = pd.read_csv('data/two_phase_falling_sphere/exp_data_for_falling_sphere_water_air.csv')
```

```python
data['x1'] = data['x']/5
data['y1'] = ((data[' y'])/2)+0.9
```

```python
linienstaerke = 2
MarkerGroesse = 8
plt.figure(figsize=(10, 6))
plt.plot(par1[:, 0], par1[:, 3], '--',
         linewidth=linienstaerke,color ='red', markersize=MarkerGroesse)
plt.xlim([0, 3])
plt.plot(data['x1'],data['y1'],'--',linewidth=linienstaerke,color = 'black', markersize=MarkerGroesse)
plt.xlabel('Dimentsionless time',fontsize = 20)
plt.ylabel('Position',fontsize = 20)
#plt.title('Comparison of the z-position of a particle', fontsize=15)
plt.legend(['Resolved CFD-DEM', 'Pathak (2016)'], frameon=False,fontsize = 14)
plt.savefig('bouncing_sphere_plot.png');
```
