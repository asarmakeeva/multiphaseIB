```python
import numpy as np
import matplotlib.pyplot as plt

# load data from simulation
par1 = np.loadtxt('../../DEM/post/position_particle_1.txt')
#par2 = np.loadtxt('../../DEM/post/position_particle_2.txt')
par1vel = np.loadtxt('../../DEM/post/velocity_particle_1.txt')
#par2vel = np.loadtxt('../../DEM/post/velocity_particle_2.txt')
```

```python
import pandas as pd
data = pd.read_csv('plot-data.csv')
#plt.plot(data['x'],data[' y']);
```

```python
# compute offset
#offset = 5 - par1[0, 3]

# plot the data
linienstaerke = 2
MarkerGroesse = 8
plt.figure(1)
#plt.plot(par1[:, 0], offset + par1[:, 3], '*',
plt.plot(par1[:, 0], par1[:, 3], '-',
         linewidth=linienstaerke, markersize=MarkerGroesse)
plt.plot(data['x'],data[' y'])
plt.xlabel('time (s)')
plt.ylabel('position (cm)')
plt.title('Comparison of the z-position of two particle', fontsize=15)
plt.legend(['following particle', 'leading particle'])
#plt.axis([0, 0.25, 0, 6])
plt.tick_params(labelsize=14)
plt.show()

plt.figure(2)
plt.plot(par1vel[:, 0], par1vel[:, 3], '-', 
         linewidth=linienstaerke, markersize=MarkerGroesse)

plt.xlabel('time (s)')
plt.ylabel('position (cm)')
plt.title('Comparison of the settling velocity of a particle', fontsize=15)
plt.legend(['following particle', 'leading particle'])
#plt.axis([0, 0.25, -20, 0])
plt.tick_params(labelsize=14)
#plt.show()
```

```python
data['x1'] = data['x']/6.216
data['y1'] = ((data[' y'])/2)+0.9

```

```python
plt.plot(par1[:, 0]*0.8, par1[:, 3], '--',
         linewidth=linienstaerke,color ='red', markersize=MarkerGroesse)
plt.xlim([0, 2.3])
plt.plot(data['x1'],data['y1'],'--',linewidth=linienstaerke,color = 'black', markersize=MarkerGroesse)
plt.xlabel('dimentsionless time')
plt.ylabel('position')
plt.title('Comparison of the z-position of a particle', fontsize=15)
plt.legend(['Resolved CFD-DEM', 'Pathak[2]'])
plt.savefig('plot.png');
```
